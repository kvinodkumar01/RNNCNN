```python
# Importing Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# Importing Data from CSV file

dataset = pd.read_csv('D:\RNNCNN\Google_Stock_Price_Train.csv', date_parser = True)
dataset
```

```python
dataset_train = dataset[dataset['Date']<'2021-01-01'].copy()
dataset_train
```

```python
dataset_test = dataset[dataset['Date']>='2021-01-01'].copy()
dataset_test
```

```python
training_data = dataset_train.drop(['Date', 'Adj Close'], axis =1)
training_data.head()
```

```python
# Scaling
scaler =MinMaxScaler()
training_data = scaler.fit_transform(training_data)
training_data
```

```python
# Creating a datastructure with 60 timesteps and 1 output
x_train = []
y_train = []
for i in range(60,training_data.shape[0]):
    x_train.append(training_data[i-60:i])
    y_train.append(training_data[i,0])
x_train, y_train = np.array(x_train), np.array(y_train)

#Reshaping
#x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
```

```python
x_train.shape, y_train.shape
```

```python
# Building LSTM
# Import Libraries and Packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
regressor = Sequential()
```

```python
# Adding the first LSTM Layer and some Dropout regularisation
regressor.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape = (x_train.shape[1],5)))
regressor.add(Dropout(0.2))
```

```python
# Adding Second LSTM Layer and some Dropout regularisation
regressor.add(LSTM(units=60, activation='relu', return_sequences=True))
regressor.add(Dropout(0.3))
```

```python
# Adding Third LSTM Layer and some Dropout regularisation
regressor.add(LSTM(units=80, activation='relu', return_sequences=True))
regressor.add(Dropout(0.4))
```

```python
# Adding Fourth LSTM Layer and some Dropout regularisation
regressor.add(LSTM(units=120, activation='relu'))
regressor.add(Dropout(0.5))
```

```python
# Adding output layer
regressor.add(Dense(units=1))

regressor.summary()
```

```python
regressor.compile(optimizer ='adam', loss = 'mean_squared_error')

regressor.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python
# Prepare test dataset
past_60_days = dataset_test.tail(60)
past_60_days
```

```python
df = past_60_days.append(dataset_test, ignore_index = True)
df = df.drop(['Date', 'Adj Close'], axis=1)
df.head()
```

```python
inputs = scaler.transform(df)
inputs
```

```python
x_test =[]
y_test =[]

for i in range(60, inputs.shape[0]):
    x_test.append(inputs[i-60:i])
    y_test.append(inputs[i,0])

x_test, y_test = np.array(x_test), np.array(y_test)
x_test.shape, y_test.shape
```

```python
y_pred = regressor.predict(x_test)
scaler.scale_
```

```python
scale = 1/5.63345301e-04
scale
```

```python
y_pred = y_pred*scale
y_test = y_test*scale
```

```python
# Result Visualization
plt.figure(figsize=(14,5))
plt.plot(y_test, color ='red', label = 'Real Stock Price')
plt.plot(y_pred, color ='blue', label = 'Predicted Stock price')
plt.title('Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
